```python
import re
import string
def clean_text(txt):
  texts=[]
  for sentence in txt:
    sentence1 = ("".join(word for word in sentence if word not in string.punctuation))
    texts.append(' '.join(nltk.stem.WordNetLemmatizer().lemmatize(word) for word in nltk.word_tokenize(sentence1) if len(word)>1))
  return texts
```

```python
def create_vocab(text):
  unique_words = set()

  for sentence in text:
    for word in sentence.split(' '):
      if len(word)>2:
        unique_words.add(word.lower())


  vocab = {}
  for index, word in enumerate(sorted(list(unique_words))):
    vocab[word] = index

  return vocab
```

```python
clean_corpus = clean_text(corpus)
```

```python
print(clean_corpus)
```

```python
vocabulary = create_vocab(clean_corpus)
```

```python
text = " ".join(sentence for sentence in clean_corpus)
word_list = nltk.word_tokenize(text)
```

# 1) Apply CountVectorizer on dataset to get Top 50 words by count.

```python
from nltk.probability import FreqDist
fdist = FreqDist(word_list)
comm = fdist.most_common(50)
```

```python
words = [ele[0] for ele in comm]
freq = [ele[1] for ele in comm]
pd.DataFrame({"Words":words, "Frequency":freq})
```

# 2) Apply TFIDF Vectorizer on dataset to get Top 50 words with high tfidf score and create word cloud for it.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()

tfidf = vec.fit_transform(clean_corpus)
```

```python
pd.DataFrame(tfidf.toarray(), columns = vec.get_feature_names_out())
```

```python
# map feature names to sum of vector array
tfidf_dict = dict(zip(vec.get_feature_names_out(), tfidf.toarray().sum(axis=0)))
# sort in descending order
tfidf_dict = dict(sorted(tfidf_dict.items(), key=lambda x: x[1], reverse=True))
```

```python
commtfidf = list(tfidf_dict.items())[:50]
```

```python
word = [ele[0] for ele in commtfidf]
frequ = [ele[1] for ele in commtfidf]
pd.DataFrame({"Words":word, "Frequency":frequ})
```
