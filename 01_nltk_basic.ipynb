```python
words = []
for word in nltk.word_tokenize(text):
  words.append(word)

print(words[:5])
```

```python
tags = []
for sent in nltk.sent_tokenize(text):
  tags.append(nltk.pos_tag(nltk.word_tokenize(sent)))
print(tags[8])
```

```python
from nltk import chunk
tree = chunk.ne_chunk(tags[8])
print(tree)
```

```python
# Stemming
from nltk.stem.snowball import SnowballStemmer
snowball = SnowballStemmer('english')

stemmed = []

for word in nltk.word_tokenize(text):
  stemmed.append(snowball.stemmer.stem(word))

print(stemmed)
```

```python
# Lemmatizing
from nltk.stem.wordnet import WordNetLemmatizer
lemma = []

for word in nltk.word_tokenize(text):
  lemma.append(WordNetLemmatizer().lemmatize(word))

print(lemma)
```

```python
from nltk.corpus import stopwords

# Empty list to Store words with no punctuation
words_wo_punc = []

# Removing punctuations
for w in nltk.word_tokenize(text):
  if w.isalpha():
    words_wo_punc.append(w.lower())

# printing the text without punctuations
print(words_wo_punc[10:40])

# Removing stopwords
stop_words = set(stopwords.words('english'))
filtered_sentence = [w for w in words_wo_punc if not w in stop_words]

print(filtered_sentence[10:40])
```
